```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Load ค่า Root Mean Square Error (RMSE) และค่า R-squared ที่ได้จากการประเมิน Linear Regression และ Support Vector Regression (SVR):

```python
RMSE_LR = float(open('./eval_results/RMSE_LR.txt').read())
RMSE_SVR = float(open('./eval_results/RMSE_SVR.txt').read())
R2_LR = float(open('./eval_results/R2_LR.txt').read())
R2_SVR = float(open('./eval_results/R2_SVR.txt').read())
```

แสดงกราฟเปรียบเทียบค่า RMSE และ R-squared ระหว่าง Linear Regression และ Support Vector Regression (SVR):

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))

# Plot RMSE values
sns.barplot(x = ['LinRegress','SVR'],
            y = [RMSE_LR, RMSE_SVR], ax=ax[0])
ax[0].set(ylabel='RMSE score', ylim=(40,50))
ax[0].set_title('RMSE (less is better)', fontsize=15)

# Plot R-squared values
sns.barplot(x = ['LinRegress','SVR'],
            y = [R2_LR, R2_SVR], ax=ax[1])
ax[1].set(ylabel='R2 score', ylim=(0.4,0.6))
ax[1].set_title('R2 (more is better)', fontsize=15)

plt.show()
```

```python
print('Linear Regression:')
print('Root Mean Square Error (RMSE) =', RMSE_LR)
print('R2 Score =', R2_LR)

print('\nSupport Vector Regression (SVR):')
print('Root Mean Square Error (RMSE) =', RMSE_SVR)
print('R2 Score =', R2_SVR)
```

---

**ค่า Root Mean Square Error (RMSE)**  
บ่งบอกถึงผลต่างระหว่างค่าจริงกับค่าทำนาย ยิ่งต่ำยิ่งบ่งบอกว่า model fit กับชุดข้อมูลได้ดี:
1) สำหรับ Linear Regression มีค่า RMSE เท่ากับ 45.93 USD
2) สำหรับ Support Vector Regression (SVR) มีค่า RMSE เท่ากับ 46.41 USD

นั่นคือ ทั้งสอง models fit กับชุดข้อมูลได้แทบไม่ต่างกัน ความแม่นยำใกล้เคียงกัน
